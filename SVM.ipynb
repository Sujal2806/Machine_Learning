```python
#SVM
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
```

```python
#generating sysnthetic data for illustration
np.random.seed(42)
X=np.random.randn(20,2)# 20 data points with feature 2
y=np.where(X[:,0]+X[:,1]>0,1,-1)#creataing label based on linear decision
```

```python
#fit the svm model with a linear kernel and a large C value for hard margin
C_value=1e6#use a large value to approx a hard margin svm
clf=svm.SVC(kernel="linear",C=C_value)
clf.fit(X,y)
```

```python
#plotting the decision boundary and support vectors
plt.scatter(X[:,0],X[:,1],c=y,cmap=plt.cm.Paired)
ax=plt.gca()
```

```python
#plot the decision boundary
xlim=ax.get_xlim()
ylim=ax.get_ylim()

#create grid to evaluate model
xx,yy=np.meshgrid(np.linspace(xlim[0],xlim[1],50),
                  np.linspace(ylim[0],xlim[1],50))
```

```python
#plot decision func for each datapoint in the grid
Z=clf.decision_function(np.c_[xx.ravel(),yy.ravel()])
Z=Z.reshape(xx.shape)
```

```python
#plot decision boundary and margins
ax.contour(xx,yy,Z,colors='k',levels=[-1,0,1],alpha=0.5,
           linestyles=['--','-','--'])
```

```python
#highlight support vectors
ax.scatter(clf.support_vectors_[:,0],clf.support_vectors_[:,1],s=100,
           linewidth=1,facecolors='none',edgecolors='k')
plt.title("Maximum margin classifier with svm")
plt.show()
```

```python
#svm
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm

#generating sysnthetic data for illustration
np.random.seed(42)
X=np.random.randn(20,2)# 20 data points with feature 2
y=np.where(X[:,0]+X[:,1]>0,1,-1)#creataing label based on linear decision

#fit the svm model with a linear kernel and a large C value for hard margin
C_value=1e6#use a large value to approx a hard margin svm
clf=svm.SVC(kernel="linear",C=C_value)
clf.fit(X,y)

#plotting the decision boundary and support vectors
plt.scatter(X[:,0],X[:,1],c=y,cmap=plt.cm.Paired)
ax=plt.gca()

#plot the decision boundary
xlim=ax.get_xlim()
ylim=ax.get_ylim()

#create grid to evaluate model
xx,yy=np.meshgrid(np.linspace(xlim[0],xlim[1],50),
                  np.linspace(ylim[0],xlim[1],50))

#plot decision func for each datapoint in the grid
Z=clf.decision_function(np.c_[xx.ravel(),yy.ravel()])
Z=Z.reshape(xx.shape)

#plot decision boundary and margins
ax.contour(xx,yy,Z,colors='k',levels=[-1,0,1],alpha=0.5,
           linestyles=['--','-','--'])
#highlight support vectors
ax.scatter(clf.support_vectors_[:,0],clf.support_vectors_[:,1],s=100,
           linewidth=1,facecolors='none',edgecolors='k')
plt.title("Maximum margin classifier with svm")
plt.show()
```
